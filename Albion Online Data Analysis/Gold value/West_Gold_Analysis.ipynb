```python
import pandas as pd

# Read the .csv file
df = pd.read_csv('WestPrices10-25-23.csv')
# Formatting to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
# Changes the name of the collumn and drops the old one (could use some work)
df['date'] = df['timestamp']
df = df.drop(columns=['timestamp'])
# Sets the new collumn as index
df.set_index('date', inplace=True)
df
```

```python
# Resampling by month
df = df.resample('M').mean()
# Adds the 'pct_change' column 
df['pct_change'] = df['price'].pct_change()
# Drops NaN values in the pct_change
df = df.dropna()
# Rounds the 'pct_change' collumn to 2 decimal houses
df['pct_change'] = (df['pct_change'] * 100).round(2)
df
```

```python
# Which of the months has the highest difference in pct_change (When to invest)
monthly_data = df.groupby(df.index.month)['pct_change'].median()

monthly_data
```

```python
# Remove the pre-free-to-play data

# Create a boolean mask for the date range to remove
mask = ~((df.index >= '2018-01-31') & (df.index <= '2019-10-31'))

# Filter the DataFrame using the boolean mask
df = df.loc[mask]

df
```

```python
# Dump the .csv file to the folder

monthly_data.to_csv('whenToInvest.csv')
```
